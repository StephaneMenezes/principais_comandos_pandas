**FILTRANDO VALORES**

Existem dois métodos principais para resgatar e filtrar valores em um DataFrame, cada um com suas características distintas.
No lugar de tentar explica-lós, algo que é muito confuso para quem nunca utilizou nenhum deles, vamos ver exemplos da utilização: 

**LOC**: O LOC PODE FILTRAR LINHAS ATRÁVES DE ALGUMA INFORMÇÃO CONTIDA NA LINHA: 

```python
df.loc[201305055424986]
```

```python
#Filtrando para apenas colunas especificas:
df.loc[201305055424986,["ocorrencia_pais","ocorrencia_tipo"]]
```

```python
# Filtrando para um tipo especifico:

df.loc[df.ocorrencia_tipo=="FALHA DO MOTOR EM VOO"].head()
```
